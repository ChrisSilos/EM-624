```python
# Import libraries
import numpy as np
import pandas as pd

file = open('citi_bike.txt') # Read the text file
text_data = [] # Define a list to hold the information from the text file

for line in file: # Read each line in the file
    parts = line.strip().split() # Strip the white space and create a list of items
    text_data.append(parts) # Append that item's list to the data list
```

```python
def print_details(data):

    daily_miles = [float(line[3]) for line in data] # Create a list to hold the daily miles for each line
    pass_purchases = [int(line[7]) for line in data] # Create a list to hold the daily pass purchases for each line
    trips = [int(line[1]) for line in data] # Create a list to hold the cumulative trips for each line
    dates = [line[0] for line in data]

    for i in range(len(trips)):
        for j in range(i + 1, len(trips)):
            if(trips[i] > trips[j]):
                temp_trip = trips[i]
                temp_date = dates[i]

                trips[i] = trips[j]
                dates[i] = dates[j]

                trips[j] = temp_trip
                dates[j] = temp_date


    average_daily_miles = np.array(daily_miles).mean() # Calculate the mean daily miles for the dataset
    total_passes = np.array(pass_purchases).sum() # Calculate the sum of the total passes for the dataset

    print(f'\nThe following data is from {data[0][0]} to {data[-1][0]}') # Print the range of dates for the dataset
    print(f'\nAverage daily miles: {average_daily_miles:.2f} mi') # Print the average daily miles
    print(f'\nTotal number of passes purchased: {total_passes} passes') # Print the total number of passes purchased
    print(f'\nThree days with highest number of trips: {dates[-3:]}')
    print(f'\nThree days with lowest number of trips: {dates [:3]}')
```

```python
print_details(text_data)
```

```python
file = open('citi_bike.csv') # Read the text file
csv_data = [] # Define a list to hold the information from the text file

for line in file: # Read each line in the file
    parts = line.strip().split(',') # Strip the white space and create a list of items
    csv_data.append(parts) # Append that item's list to the data list
```

```python
print_details(csv_data)
```

```python
# Create list to hold column titles
headers = ['Date', 'Trips over last 24h', 'Cumulative trips', 'Miles traveled today','Miles traveled to date', 'Total Annual Members', 'Annual Member Sign-ups', '24hr passes purchased','7 day passes purchased']

txt_df = pd.DataFrame(text_data, columns=headers) # Create a dataframe for the text data
csv_df = pd.DataFrame(csv_data, columns=headers) # Create a dataframe for the csv data
final_df = pd.concat([csv_df,txt_df], ignore_index=True) # Concatenate the dataframes
```
